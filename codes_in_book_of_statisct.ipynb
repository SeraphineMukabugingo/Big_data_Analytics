```python
import matplotlib.pyplot as plt

# Define abline function
def abline(ax, b, m):
    """
    Add a line with slope m and intercept b to the given Axes object `ax`.
    """
    xlim = ax.get_xlim()
    x_vals = [xlim[0], xlim[1]]
    y_vals = [m * x + b for x in x_vals]
    ax.plot(x_vals, y_vals, color='red', linestyle='--')
    ax.set_xlim(xlim)  # Reset x-limits to original

# Create a plot
fig, ax = plt.subplots()
ax.scatter([1, 2, 3], [2, 4, 5])  # Example scatter points

# Add the line y = 1.5x + 1
abline(ax, b=1, m=1.5)

# ✅ Required to display the plot!
plt.show()
```

```python
def boot_SE(func, D, n=None, B=1000, seed=0):
    """
    Estimate the standard error of a statistic using bootstrap.
    
    Parameters:
    - func: a function that takes (D, idx) and returns a statistic
    - D: the full DataFrame
    - n: sample size for each bootstrap draw (default = len(D))
    - B: number of bootstrap samples (default = 1000)
    - seed: random seed for reproducibility

    Returns:
    - Estimated standard error of the statistic
    """
    rng = np.random.default_rng(seed)  # Create random number generator
    first_, second_ = 0.0, 0.0         # Accumulators for mean and mean squared
    n = n or D.shape[0]                # Use full dataset size if n is not provided

    for _ in range(B):
        idx = rng.choice(D.index, n, replace=True)  # Bootstrap sample indices
        value = func(D, idx)                        # Compute the statistic
        first_ += value
        second_ += value**2

    # Standard error formula: sqrt(E[X^2] - (E[X])^2)
    return np.sqrt(second_ / B - (first_ / B)**2)
```

```python
alpha_SE = boot_SE(alpha_func ,
                   Portfolio ,
                   B=1000,
                   seed=0)
alpha_SE
```
